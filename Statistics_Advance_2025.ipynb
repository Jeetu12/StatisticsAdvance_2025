### **1. What is a random variable in statistics, and how is it used in probability theory?**

**Role**: Data Scientist, Data Analyst
**Companies**: Google, Oracle
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
A **random variable** is a numerical outcome of a random phenomenon. In probability theory, it assigns a real number to each outcome in a sample space. Random variables allow us to **quantify uncertainty** and **model real-world phenomena**.
There are two types:

* **Discrete Random Variables**: take on a countable number of values (e.g., number of heads in 3 coin tosses).
* **Continuous Random Variables**: take on an infinite number of values within a given range (e.g., height, weight).

They form the foundation for defining **probability distributions**, calculating **expectations**, and conducting **inference**.

### **3. How do you calculate the expected value of a random variable?**

**Role**: Data Scientist, Data Analyst
**Companies**: Google, Oracle
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
The **expected value (mean)** of a random variable gives the long-run average outcome of a random process.

* For **Discrete Random Variable**:

  $$
  E[X] = \sum_{i} x_i \cdot P(x_i)
  $$

* For **Continuous Random Variable**:

  $$
  E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
  $$

Where:

* $x_i$: possible value
* $P(x_i)$: probability of $x_i$
* $f(x)$: probability density function

* In **regression models**, outcomes are treated as random variables influenced by inputs.

They allow analysts to calculate probabilities, expectations, variances, and guide decisions under uncertainty.

### **6. Explain the difference between probability density function (PDF) and probability mass function (PMF).**

**Role**: Data Analyst, Business Analyst
**Companies**: Amazon, Zomato
**Difficulty**: ‚≠ê‚≠ê

**Answer**:

| Feature        | PMF                                | PDF                                |
| -------------- | ---------------------------------- | ---------------------------------- |

| Definition     | $P(X = x)$ gives exact probability | $f(x)$ gives probability *density* |

> **Note**: For continuous distributions, the probability of a specific value is **zero**, so we talk about probability in intervals using PDF.

### **7. What is the cumulative density function (CDF), and how is it related to PDF and PMF?**

**Role**: Data Science, Data Analytics
**Companies**: Microsoft, Swiggy
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
The **Cumulative Distribution Function (CDF)** gives the probability that a random variable is **less than or equal to a value** $x$:

$$
F(x) = P(X \leq x)
$$

* For **Discrete Variables**: $F(x) = \sum_{t \leq x} P(t)$
* For **Continuous Variables**: $F(x) = \int_{-\infty}^{x} f(t) dt$

**Relation**:

* **PDF is the derivative of the CDF** for continuous variables:

  $$
  f(x) = \frac{dF(x)}{dx}
  $$

* **CDF is the summation of PMF** for discrete variables.

### **13. How is a probability density function (PDF) used in continuous probability distributions?**

**Role**: Data Scientist, ML Engineer
**Companies**: Flipkart, Google
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
A **PDF** describes the **relative likelihood** of a continuous random variable taking on a specific value.

Key points:

* $P(a < X < b) = \int_{a}^{b} f(x)\,dx$
* The **total area** under the PDF curve = 1
* The value of the PDF at a point is **not a probability**, but **probability density**

Example: Normal distribution

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
$$

### **14. Explain how the probability mass function (PMF) is used for discrete distributions.**

**Role**: Data Analyst, Data Scientist
**Companies**: Amazon, Swiggy
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
A **PMF** assigns probabilities to each **discrete outcome** of a random variable.

$$
P(X = x_i) = p_i
\quad \text{where} \quad \sum_i p_i = 1
$$

Examples:

* Tossing a coin: P(H) = 0.5, P(T) = 0.5
* Rolling a die: P(1) = ... = P(6) = 1/6

The PMF allows:

* **Exact probability calculations**
* Construction of **discrete distribution tables**

### **16. What is a discrete uniform distribution, and when is it used?**

**Role**: Data Scientist, Data Analyst
**Companies**: Flipkart, TCS
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
A **Discrete Uniform Distribution** is a probability distribution where each **outcome in a finite set** is **equally likely**.

#### Formula:

If a random variable $X$ takes $n$ discrete values, each with equal probability:

$$
P(X = x_i) = \frac{1}{n}, \quad \text{for } i = 1, 2, ..., n
$$

#### Use cases:

* Rolling a **fair die**
* Picking a **random card** from a deck
* **Random sampling** with replacement

> üîç Used when there's **no bias or preference** among outcomes.

### **21. What is the Bernoulli distribution, and what are its properties?**

**Role**: Data Scientist, Data Analyst
**Companies**: Flipkart, Ola
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
A **Bernoulli distribution** models a **single trial** that results in either **success (1)** or **failure (0)**, with probability $p$ of success.

#### Properties:

* Only **two outcomes**: 1 (success), 0 (failure)
* Mean: $\mu = p$
* Variance: $\sigma^2 = p(1 - p)$
* PMF:

  $$
  P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0, 1\}
  $$

> üîç Used in binary events like pass/fail, win/lose, click/no-click.

---

### **22. Explain how a Bernoulli trial is used in binary classification problems.**

**Role**: ML Engineer, Data Scientist
**Companies**: Google, Swiggy
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
A **Bernoulli trial** simulates **yes/no** outcomes in binary classification, such as:

* Spam or not spam
* Disease present or not

In a **binary classifier** (like logistic regression), the predicted probability $p$ of the positive class can be modeled using a Bernoulli distribution:

$$
P(Y = 1) = p, \quad P(Y = 0) = 1 - p
$$

This forms the basis of **loss functions** like binary cross-entropy.

---

### **23. How does the Bernoulli distribution relate to the binomial distribution?**

**Role**: Data Scientist, Data Analyst
**Companies**: Flipkart, Ola
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
The **Binomial distribution** is the **sum of independent Bernoulli trials**.

If $X_1, X_2, ..., X_n \sim \text{Bernoulli}(p)$, then:

$$
S = X_1 + X_2 + \dots + X_n \sim \text{Binomial}(n, p)
$$

* Bernoulli: One trial
* Binomial: $n$ trials

> üîÅ Binomial = Repeated Bernoulli

### **25. What is the binomial distribution, and how does it differ from the Bernoulli distribution?**

**Role**: Data Scientist, Business Analyst
**Companies**: Amazon, Flipkart
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
A **Binomial distribution** models the **number of successes in $n$ independent Bernoulli trials**, each with probability $p$.

#### PMF:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

**Differences**:

| Feature  | Bernoulli  | Binomial    |
| -------- | ---------- | ----------- |
| Trials   | 1          | $n$         |
| Values   | 0 or 1     | 0 to $n$    |
| Mean     | $p$        | $np$        |
| Variance | $p(1 - p)$ | $np(1 - p)$ |

---

### **26. Explain how to calculate the probability of multiple events in a binomial distribution.**

**Role**: Data Analyst, Data Scientist
**Companies**: Google, Zomato
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
To calculate $P(X = k)$ for $k$ successes in $n$ trials:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

**Example**: Flip a coin 5 times, probability of 3 heads:

$$
P(X = 3) = \binom{5}{3} (0.5)^3 (0.5)^2 = 10 \times 0.125 \times 0.25 = 0.3125
$$

### **27. What is the binomial coefficient, and why is it important in the binomial distribution?**

**Role**: Data Scientist, Data Analyst
**Companies**: Flipkart, Ola
**Difficulty**: ‚≠ê‚≠ê

**Answer**:
The **binomial coefficient** $\binom{n}{k}$ represents the **number of ways** to choose $k$ successes from $n$ trials.

$$
\binom{n}{k} = \frac{n!}{k!(n - k)!}
$$

It accounts for the **combinatorics** of different success arrangements in the binomial PMF.

**Answer**:
A **continuous uniform distribution** is where all outcomes in a continuous range $[a, b]$ are **equally likely**.

### **34. Explain how to calculate the probability of an event in a continuous uniform distribution.**

**Role**: Data Scientist, Business Analyst
**Companies**: Microsoft, Ola
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
For a continuous uniform distribution between $a$ and $b$, the probability that $X$ lies between $x_1$ and $x_2$ is:

$$
P(x_1 \leq X \leq x_2) = \frac{x_2 - x_1}{b - a}
$$

‚úÖ **Example**: If $X \sim U(10, 20)$, find $P(12 \leq X \leq 15)$:

$$
P = \frac{15 - 12}{20 - 10} = \frac{3}{10} = 0.3
$$

### **45. What is the central limit theorem, and why is it important in statistics?**

**Role**: Data Science, Data Analyst, Machine Learning Engineer
**Companies**: Microsoft, Zomato
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
The **Central Limit Theorem (CLT)** states that the **sampling distribution** of the **sample mean** approaches a **normal distribution** as the **sample size becomes large**, regardless of the original population distribution.

Mathematically:
If $X_1, X_2, ..., X_n$ are i.i.d. with mean $\mu$ and standard deviation $\sigma$, then:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i \approx N\left(\mu, \frac{\sigma}{\sqrt{n}}\right)
$$

**Importance**:

* Enables the use of **normal approximation** for many statistical methods
* Forms the foundation of **hypothesis testing**, **confidence intervals**, and **z-tests**

---

### **46. How does the central limit theorem apply to large sample sizes?**

**Role**: Data Science, Data Analyst, Machine Learning Engineer
**Companies**: Microsoft, Zomato
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
As the sample size **$n$** increases:

* The distribution of the **sample mean** becomes increasingly **normal**.
* The **standard error** decreases:

  $$
  \text{Standard Error} = \frac{\sigma}{\sqrt{n}}
  $$

**Rule of Thumb**:
Sample size $n \geq 30$ is generally sufficient for the CLT to hold, even if the population is skewed.

---

### **47. Explain the role of the central limit theorem in hypothesis testing.**

**Role**: Data Science, Data Analyst, Machine Learning Engineer
**Companies**: Microsoft, Zomato
**Difficulty**: ‚≠ê‚≠ê‚≠ê

**Answer**:
CLT justifies the use of **z-tests** and **t-tests** by ensuring that the distribution of sample means is approximately **normal** for large $n$.

**In hypothesis testing**:

* The null hypothesis assumes a population mean
* We test the sample mean using the normal approximation from CLT
* Enables calculation of p-values and critical regions

### 13. **What is a confidence interval, and why is it important in inferential statistics?**

A **confidence interval (CI)** is a range of values derived from sample data that is likely to contain the **true population parameter** (like the mean or proportion) with a certain level of confidence.

For example, a 95% CI means we are 95% confident the true parameter lies within this range.

**Importance:**

* Helps **quantify uncertainty** in estimates.
* Gives a **range** rather than a single point estimate.
* Critical in **decision-making** and **hypothesis testing**.

---

### 14. **Explain the relationship between confidence intervals and the margin of error.**

The **confidence interval** is computed as:

$$
\text{CI} = \text{point estimate} \pm \text{margin of error}
$$

* **Margin of Error (ME)** is the amount added/subtracted from the estimate to create the range.
* ME depends on:

  * **Confidence level** (e.g., 95%)
  * **Sample size** (larger size ‚Üí smaller ME)
  * **Standard deviation** or **standard error**

**Higher confidence** ‚áí **larger margin of error**, and **vice versa**.

---

### 15. **How do you interpret a 95% confidence interval in statistical analysis?**

If we compute a 95% CI for the population mean as **\[45, 55]**, we interpret it as:

> "We are 95% confident that the true population mean lies between 45 and 55."

It **does not mean** there's a 95% chance that the true value is in the interval for this particular case. Instead, it means that if we repeated the sampling process many times, **95% of the computed intervals would contain the true mean**.

### 16. **How can you calculate confidence intervals in Python using `scipy.stats`?**

### 17. **What is the Chi-square test, and when is it used in statistical analysis?**

The **Chi-square test** is a **non-parametric test** used to examine the relationship between **categorical variables** or assess how well observed data fits expected data.

There are two main types:

* **Chi-square test of independence**: Checks if two categorical variables are independent.
* **Chi-square goodness-of-fit test**: Checks if sample data matches a theoretical distribution.

**Use case example**: Testing if gender is independent of product preference in a survey.

---

### 18. **Explain how to perform a Chi-square test of independence.**

Steps:

1. **Create a contingency table** (e.g., rows = gender, columns = product preference).
2. **Compute expected frequencies** assuming independence.
3. **Use the formula**:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where $O_i$ = observed frequency, $E_i$ = expected frequency.

4. Compare the test statistic to the **Chi-square distribution** with appropriate **degrees of freedom (df)**:

$$
df = (rows - 1) \times (columns - 1)
$$

5. Make a decision based on the **p-value** and your significance level (e.g., 0.05).

---

### 19. **How is the Chi-square distribution related to categorical data analysis?**

* The **Chi-square distribution** is used as the reference distribution for the Chi-square test.
* It applies when testing **frequencies** or **proportions** in **categorical data**.
* It is **right-skewed** and **non-negative**.
* As the **degrees of freedom increase**, the distribution becomes more symmetric.

---


### 21. **What is Bayes‚Äô theorem, and how is it applied in data science?**

**Bayes‚Äô theorem** describes the probability of an event, based on prior knowledge of conditions that might be related to the event. It is a fundamental tool in **probabilistic reasoning** and **machine learning**.

**Formula:**

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Where:

* $P(A|B)$: Posterior probability
* $P(B|A)$: Likelihood
* $P(A)$: Prior probability
* $P(B)$: Marginal probability

**Applications in data science:**

* **Spam filtering** (e.g., Naive Bayes)
* **Medical diagnosis**
* **Predictive modeling**
* **Recommendation systems**

#### In **medical diagnostics**:

Let **A** = Has disease, **B** = Tests positive.
Bayes' calculates the chance of disease **after** a positive test, considering false positives and prevalence.

### 23. **What is the difference between prior, likelihood, and posterior probabilities in Bayes‚Äô theorem?**

| -------------- | --------------------------------------------------------------------------------------- |

### 26. **How is the Chi-square goodness-of-fit test performed?**

Where:

4. Compare the $\chi^2$ value with the critical value from the Chi-square distribution table.
5. Reject or fail to reject $H_0$.

### 29. **What is the F-distribution, and how is it used in statistical analysis?**

The **F-distribution** is a continuous probability distribution that arises frequently in ANOVA and regression analysis when comparing **two variances**.

**Key uses:**

* Comparing **variances** between two populations.
* Testing the **overall significance** in multiple regression.
* Performing **ANOVA (Analysis of Variance)**.

It is **right-skewed** and depends on two degrees of freedom: $df_1$ (numerator) and $df_2$ (denominator).

---

### 30. **Explain the relationship between the F-distribution and analysis of variance (ANOVA).**

In **ANOVA**, the F-distribution is used to compare **between-group variance** to **within-group variance**:

$$
F = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}}
$$

* If the group means are equal (null hypothesis), this ratio should be close to 1.
* A **larger F-value** implies that the group means are different, indicating a statistically significant result.

---

### 31. **What are the properties of the F-distribution in hypothesis testing?**

* **Non-negative**: F-values are always ‚â• 0.
* **Right-skewed**: Tail extends to the right.
* Depends on **two degrees of freedom**.
* **Mean > 1** for small degrees of freedom.
* Used for **variance comparison** and **ANOVA**.

---

### 32. **How do you calculate the F-statistic in Python using `scipy.stats`?**

Example: Compare variances of two datasets

```python
import scipy.stats as stats
import numpy as np

# Sample data
group1 = [12, 14, 15, 16, 17]
group2 = [22, 25, 26, 23, 24]

# Calculate variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# Compute F-statistic
f_stat = var1 / var2
dof1 = len(group1) - 1
dof2 = len(group2) - 1

# Compute p-value
p_value = 1 - stats.f.cdf(f_stat, dfn=dof1, dfd=dof2)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

### 33. **What is an F-test, and how does it differ from other statistical tests?**

An **F-test** is used to:

**Difference from other tests:**

* Unlike the **T-test** (used for means), the **F-test** compares **variances**.
* Used when **more than two groups** or multiple variables are involved.

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Data
data = {
    'score': [23, 21, 22, 24, 20, 30, 31, 29, 28, 32, 40, 39, 41, 42, 38],
    'group': ['A']*5 + ['B']*5 + ['C']*5
}

df = pd.DataFrame(data)

# One-way ANOVA
model = ols('score ~ group', data=df).fit()
anova_result = sm.stats.anova_lm(model, typ=2)

print(anova_result)
````

````python
from scipy.stats import f_oneway

# Sample data: 3 groups
group1 = [23, 21, 22, 24, 20]
group2 = [30, 31, 29, 28, 32]
group3 = [40, 39, 41, 42, 38]

# Perform ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("P-value:", p_value)
````
